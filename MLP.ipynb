**Loading the dataset**

```python
test = pd.read_csv('/content/test_sample.csv')
train = pd.read_csv('/content/train_sample.csv')
```

**Data Preprocessing**

```python
test_pixel = test.iloc[:,1:].astype('float32') / 255
test_labels = test.iloc[:,0].astype('float32')
train_pixel = train.iloc[:,1:].astype('float32') / 255
train_labels = train.iloc[:,0].astype('float32')
```

```python
train_label = to_categorical(train_labels)
test_label = to_categorical(test_labels)
```
